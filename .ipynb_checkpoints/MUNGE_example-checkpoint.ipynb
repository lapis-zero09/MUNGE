```python
from itertools import combinations
from matplotlib import pyplot as plt
from sklearn import datasets
import numpy as np
from scipy.io import arff
```

```python
np.save('./data/data.npy', iris.data)
```

```python
iris = datasets.load_iris()

features = iris.data
feature_names = iris.feature_names
targets = iris.target
```

```python
plt.figure(figsize=(12, 8))

for i, (x, y) in enumerate(combinations(range(4), 2)):
    plt.subplot(2, 3, i + 1)
    for t, marker, c in zip(range(3), '>ox', 'rgb'):
        plt.scatter(features[targets == t, x],
                    features[targets == t, y],
                    marker=marker, c=c, alpha=0.5)
        plt.xlabel(feature_names[x])
        plt.ylabel(feature_names[y])
        plt.autoscale()
        plt.grid()

plt.savefig("./img/iris.png", dpi=200)
plt.show()
```

```python
dataset, meta = arff.loadarff('./data/munge_p_0.5s_10.0.arff')
dataset = np.asarray(dataset.tolist(), dtype=np.float32)[200:]
dataset
```

```python
plt.figure(figsize=(12, 8))

for i, (x, y) in enumerate(combinations(range(4), 2)):
    plt.subplot(2, 3, i + 1)
    for _ in range(3):
        plt.scatter(dataset[:, x], dataset[:, y], marker='.', c='r')
        plt.xlabel(feature_names[x])
        plt.ylabel(feature_names[y])
        plt.autoscale()
        plt.grid()

plt.savefig("./img/munge_p_0.5s_10.0.png", dpi=200)
plt.show()
```
